```python
import pandas as pd
import plotly.express as px
from messari.messari import Messari
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

api_key = '4f11a260-28b4-4ec5-87d0-7a9cac3aa6b2'
messari = Messari(api_key)
```

# Downloading data

```python
df = pd.DataFrame()
```

```python
for i in range(1,22):
  try:
    messari_df = messari.get_all_assets(page=i, limit=500, asset_fields=['metrics'], to_dataframe=True)
    df = df.append(messari_df)
    print(len(df))
  except:
    print('no more pages')
```

```python
cleaned_df = df.reset_index()
cleaned_df = cleaned_df.applymap(lambda x: x[0] if isinstance(x, list) else x)
```

# Dashboard viz

## Market cap by sectors

```python
market_cap_sum = cleaned_df.groupby(by="metrics_misc_data_sectors")['metrics_marketcap_current_marketcap_usd'].sum().reset_index().sort_values(by='metrics_marketcap_current_marketcap_usd',ascending=False)
fig = px.bar(market_cap_sum, y='metrics_marketcap_current_marketcap_usd', x='metrics_misc_data_sectors', text='metrics_misc_data_sectors')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```

## Market cap by sectors (excluding currencies and stable coins and smartcontracts)




```python
# filtering data 

sectors = ['Stablecoins','Currencies','Smart Contract Platforms']
market_cap_sum = cleaned_df.groupby(by="metrics_misc_data_sectors")['metrics_marketcap_current_marketcap_usd'].sum().reset_index()
filtered_market_cap = market_cap_sum.loc[~market_cap_sum['metrics_misc_data_sectors'].isin(sectors)].sort_values(by='metrics_marketcap_current_marketcap_usd',ascending=False)
```

```python
fig = px.bar(filtered_market_cap, y='metrics_marketcap_current_marketcap_usd', x='metrics_misc_data_sectors', title='Market cap by sectors (excluding currencies and stable coins and smartcontracts)')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```

```python
fig = px.pie(filtered_market_cap.head(15), values='metrics_marketcap_current_marketcap_usd', names='metrics_misc_data_sectors', title='Market share')
fig.show()
```

## Gaming Market cap

```python
gaming_df = cleaned_df.loc[cleaned_df['metrics_misc_data_sectors'] == 'Gaming']
fig = px.bar(gaming_df, y='metrics_marketcap_current_marketcap_usd', x='index', text='metrics_marketcap_current_marketcap_usd')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```

```python
gaming_ts = gaming_df[['index','metrics_roi_by_year_2011_usd_percent','metrics_roi_by_year_2012_usd_percent',
                       'metrics_roi_by_year_2013_usd_percent','metrics_roi_by_year_2014_usd_percent',
                       'metrics_roi_by_year_2015_usd_percent','metrics_roi_by_year_2016_usd_percent',
                       'metrics_roi_by_year_2017_usd_percent','metrics_roi_by_year_2018_usd_percent',
                       'metrics_roi_by_year_2019_usd_percent','metrics_roi_by_year_2020_usd_percent','metrics_roi_by_year_2021_usd_percent','metrics_marketcap_current_marketcap_usd']]
```

```python
gaming_ts = gaming_df[['index','metrics_roi_by_year_2019_usd_percent','metrics_roi_by_year_2020_usd_percent','metrics_roi_by_year_2021_usd_percent','metrics_marketcap_current_marketcap_usd']]
```

```python
gaming_ts.head()
```

```python
gaming_ts['initial_marketcap'] = (gaming_ts['metrics_marketcap_current_marketcap_usd'] / gaming_ts
                                  ['metrics_roi_by_year_2021_usd_percent']) * 100
gaming_ts['2019_marketcap'] = gaming_ts['initial_marketcap'] * (gaming_ts['initial_marketcap'] * (gaming_ts['metrics_roi_by_year_2019_usd_percent']/100))
gaming_ts['2020_marketcap'] = gaming_ts['initial_marketcap'] * (gaming_ts['initial_marketcap'] * (gaming_ts['metrics_roi_by_year_2020_usd_percent']/100))
gaming_ts['2021_marketcap'] = gaming_ts['initial_marketcap'] * (gaming_ts['initial_marketcap'] * (gaming_ts['metrics_roi_by_year_2021_usd_percent']/100))
                          
```

```python
gaming_ts['log_metrics_marketcap_current_marketcap_usd'] = np.log(gaming_ts['metrics_marketcap_current_marketcap_usd'])
gaming_ts['log_2021_marketcap'] = np.log(gaming_ts['2021_marketcap'])
gaming_ts['log_2020_marketcap'] = np.log(gaming_ts['2020_marketcap'])
gaming_ts['log_2019_marketcap'] = np.log(gaming_ts['2019_marketcap'])
gaming_ts['log_initial_marketcap'] = np.log(gaming_ts['initial_marketcap'])
```

```python
fig = px.histogram(gaming_ts.sort_values('log_metrics_marketcap_current_marketcap_usd',ascending=False),x="index", y=["log_metrics_marketcap_current_marketcap_usd","log_2021_marketcap","log_2020_marketcap","log_2019_marketcap","log_initial_marketcap"],barmode='group')
fig.show()
```
